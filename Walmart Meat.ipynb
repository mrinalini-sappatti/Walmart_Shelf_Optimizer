```python
import math

def calculation_outputs(df):
    R=0.008314 #KJ/MOL
    
    degradation_rates=[]
    freshness_scores=[]
    remaining_shelf_lives = []
    risk_scores = []
    spoilage_probabilities = []
    high_risk_flags = []
    reroute_flags = []
    markdown_urgency_levels = []
    recommended_temps = []
    recommended_hums = []
    max_allowed_transit_times = []
    quality_grades = []
    action_recommendations=[]
    
    def c_to_k(c):
        return c + 273.15
    
    for idx,row in df.iterrows():
        
        product_type=str(row['Product_Type']).strip().lower()
        product_kind=str(row['Product_Kind']).strip().lower()
        
        k=0.02
        Ea=45
        alpha=0.05
        risk_temp_wt=1.3
        risk_microbe_wt=1.5
        risk_delay_wt=1.0
        risk_shelf_wt=1.5
        
        tvb_n_spoilage=40
        tba_spoilage=5
        tvb_wt=1.5
        tba_wt=1.2
        freshness_a=0.4
        freshness_b=0.6
        
        if product_type in ['poultry','meat']:
            if product_type=='poultry':
                k = 0.02
                alpha = 0.045
            else:
                k=0.012
                alpha=0.030
            Ea = 40
            risk_temp_wt = 0.4
            risk_microbe_wt = 0.3
            risk_delay_wt = 0.25
            risk_shelf_wt = 0.4
            risk_deg_wt=0.3
            risk_fresh_wt=0.35
            
            tvb_n_spoilage = 20
            tba_spoilage = 2
            tvb_wt = 0.2
            tba_wt = 0.15
            
            
        if product_kind == 'turkey':
            tvb_n_spoilage = 18
            tba_spoilage = 1.8
        elif product_kind == 'chicken':
            tvb_n_spoilage = 20
            tba_spoilage = 2
        elif product_kind == 'beef':
            tvb_n_spoilage = 40
            tba_spoilage = 5
        elif product_kind == 'pork':
            tvb_n_spoilage = 35
            tba_spoilage = 4
            
        a = 0.3
        b = 0.06
        c = 0.04
        d = -0.03
        e=0.05
        f=0.04
        g=15
        h=-0.1
        markdown_scale = 5
        reroute_delay_threshold = 12
        reroute_breach_threshold = 2
        risk_threshold = 85
        min_degradation_threshold = 0.01
        
        T_opt = c_to_k(row['Optimal_Temp_Max (°C)'])
        T_curr = c_to_k(row['Current_Temperature (°C)'])
        storage_duration = row['Storage_Duration (hr)']
        microbial_count = row['Microbial_Count (CFU/g)']
        transit_delay = row['Transit_Delay_Hours (hr)']
        cold_chain_breaches = row['Cold_Chain_Breaches (count)']
        ammonia_release = row['Ammonia_Release_Rate (ppm)']
        max_humidity_exc = row['Max_Humidity_Excursion (%)']
        priority = row['Priority_Level (1-10)']
        
        target_shelf_life = row['Target_Shelf_Life (hr)'] 
        current_humidity = row['Current_Humidity (%)']
        tvb_n = row.get('TVB-N (mg/100g)', 0)
        tba = row.get('TBA (mg MDA/kg)', 0)
        odor_score = row.get('Odor_Score (1-10)', 5)
        color_score = row.get('Color_Score (1-10)', 5)
        texture_score = row.get('Texture_Score (1-10)', 5)
        purge_loss = row.get('Purge_Loss (%)', 0)
        
        arrhenius_factor=math.exp(Ea/R*((1/T_opt)-(1/T_curr)))
        degradation_rate=k*arrhenius_factor*(1+alpha*storage_duration)
        degradation_rate=max(min(degradation_rate,10),0.01)
        
        comp_freshness=(100-7*(degradation_rate**1.1)*storage_duration-0.3*np.log1p(microbial_count)-0.08*tvb_n-0.12*tba-0.15*purge_loss-0.15*odor_score-0.1*color_score-0.1*texture_score)
        comp_freshness=max(min(comp_freshness,100),0)
        rep_freshness=row.get('Current_Freshness_Score (1-100)', np.nan)
        if not np.isnan(rep_freshness):
            freshness_score=0.5*comp_freshness+0.5*rep_freshness
        else:
            freshness_score=comp_freshness
        
        remaining_shelf_life=target_shelf_life*(freshness_score/100)*np.exp(-3*degradation_rate)   
        
        
        
        temp_dev=max(0,T_curr-T_opt)
        temp_dev=min(temp_dev*5,100)
        
        microbial_risk=min(100,microbial_count/200)
        delay_impact=min(transit_delay*100/72,100)
        tvb_n_norm=min(tvb_n/tvb_n_spoilage * 100, 100)
        tba_norm = min(tba/tba_spoilage*100 , 100)
        # Normalize degradation_rate to 0-100
        norm_degradation = 100 * (degradation_rate - 0.01) / (0.2 - 0.01)
# Invert freshness so higher freshness means lower risk
        inv_freshness = 100 - freshness_score

        if target_shelf_life > 0:
            shelf_life_factor = (1- min(remaining_shelf_life/target_shelf_life,1.0))*100
        else:
            shelf_life_factor = 100
            
        risk_score = ( risk_temp_wt*temp_dev + risk_microbe_wt*microbial_risk + risk_delay_wt*delay_impact + risk_shelf_wt * shelf_life_factor + tvb_wt*tvb_n_norm + tba_wt*tba_norm + risk_deg_wt*norm_degradation + risk_fresh_wt*inv_freshness - 35 )
        risk_score = 100/(1+np.exp(-0.05 *(risk_score -60)))
        risk_score = max(min(risk_score,100),0)
        
        logit=(a*risk_score + b*ammonia_release + c*max_humidity_exc +d*shelf_life_factor + e*tvb_n + f*tba + g* degradation_rate + h*freshness_score)
        spoilage_probability = 1/(1+math.exp(-logit))
        
        high_risk_flag = 1 if (risk_score > risk_threshold or spoilage_probability > 0.85 ) else 0
        
        if remaining_shelf_life < 24:
            reroute_flag = 0
            action_recommendation = "Immediate Sale/Disposal"
        elif (transit_delay > reroute_delay_threshold or cold_chain_breaches > reroute_breach_threshold):
            reroute_flag = 1
            action_recommendation = "Reroute"
        else:
            reroute_flag = 2
            action_recommendation = "Continue as Planned"

        

        markdown_urgency = (
                2.2 * spoilage_probability
                + 1.2 * (1 - freshness_score / 100)
                + 1.0 * (risk_score / 100)
                + 0.8 * degradation_rate
                + 0.5 * (1 - remaining_shelf_life / (target_shelf_life + 1e-9))
                + 0.3 * (priority / 10)
                 )

# Scale and clip to 1-5 range (if you want integer levels)
        markdown_urgency = np.clip(np.round(1 + 0.8 * markdown_urgency), 1, 5)

        
        recommended_temp = 0.7 * row['Optimal_Temp_Max (°C)'] + 0.3 * row['Current_Temperature (°C)']

            
        correction_factor=0.5
        target_hum = (row['Optimal_Humidity_Min (%)'] + row['Optimal_Humidity_Max (%)']) / 2
        recommended_hum = current_humidity + correction_factor * (target_hum - current_humidity)
        recommended_hum = min(max(recommended_hum, row['Optimal_Humidity_Min (%)']), row['Optimal_Humidity_Max (%)'])
        
        remaining_shelf_life_hours = remaining_shelf_life  # already in hours
        urgency_flag = remaining_shelf_life < 48  # 2 days = 48 hours
        urgency_penalty = 6  # hours (adjust as needed for your operation)

        if urgency_flag:
            adjusted_transit_delay = max(transit_delay - urgency_penalty, 0)
        else:
            adjusted_transit_delay = transit_delay

        max_transit_time = max(0.8 * remaining_shelf_life_hours - 0.5 * adjusted_transit_delay, 0)
 
        
    
        #Quality grade
        
        if freshness_score >= 85 and spoilage_probability < 0.1:
            quality_grade = 'A'
        elif freshness_score >= 70 and spoilage_probability < 0.3:
            quality_grade = 'B'
        else:
            quality_grade = 'C'
        
        degradation_rates.append(degradation_rate)
        freshness_scores.append(freshness_score)
        risk_scores.append(risk_score)
        remaining_shelf_lives.append(remaining_shelf_life)
        spoilage_probabilities.append(spoilage_probability)
        high_risk_flags.append(high_risk_flag)
        reroute_flags.append(reroute_flag)
        markdown_urgency_levels.append(markdown_urgency)
        recommended_temps.append(recommended_temp)
        recommended_hums.append(recommended_hum)
        max_allowed_transit_times.append(max_transit_time)
        quality_grades.append(quality_grade)
        action_recommendations.append(action_recommendation)
        
    df['Degradation_Rate'] = degradation_rates
    df['Freshness_Score'] = freshness_scores
    df['Risk_Score'] = risk_scores
    df['Remaining_Shelf_Life'] = remaining_shelf_lives
    df['Spoilage_Probability'] = spoilage_probabilities
    df['High_Risk_Flag'] = high_risk_flags
    df['Reroute_Recommended_Flag'] = reroute_flags
    df['Action_Recommendation']=action_recommendations
    df['Markdown_Urgency_Level'] = markdown_urgency_levels
    df['Recommended_Temperature'] = recommended_temps
    df['Recommended_Humidity'] = recommended_hums
    df['Max_Allowed_Transit_Time'] = max_allowed_transit_times
    df['Quality_Grade'] = quality_grades 
        
    return df
        
        
        
        
    
```

```python
from sklearn.model_selection import train_test_split,StratifiedKFold,cross_val_score
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import GradientBoostingClassifier,GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error,r2_score,accuracy_score
from sklearn.impute import SimpleImputer
import joblib
```

```python
x=df.iloc[:,:64]
y=df.iloc[:,64:]
```

```python
y.isnull().sum()
```

```python
numerical_features=x.select_dtypes(include=np.number).columns.tolist()
categorical_features=x.select_dtypes(exclude=np.number).columns.tolist()
```

```python
numerical_features
```

```python
categorical_features
```

```python
preprocessor=ColumnTransformer(transformers=[
    ('num',Pipeline([
        ('imputer',SimpleImputer(strategy='median')),
        ('scaler',StandardScaler())
        
    ]),numerical_features),
    
    ('cat',Pipeline([
        ('imputer',SimpleImputer(strategy='most_frequent')),
        ('encoder',OneHotEncoder(handle_unknown='ignore'))
    ]),categorical_features)
])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
y_test
```

```python


models={}
classification_indices=[5,6,7,12]
from sklearn.model_selection import KFold,GridSearchCV
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

y_train_imputed = y_train.copy()
y_test_imputed = y_test.copy()

# Impute classification columns
class_imputer = SimpleImputer(strategy='most_frequent')
for idx in classification_indices:
    col = y_train.columns[idx]
    y_train_imputed[col] = class_imputer.fit_transform(y_train[[col]]).ravel()
    y_test_imputed[col] = class_imputer.transform(y_test[[col]]).ravel()
    le = LabelEncoder()
    y_train_imputed[col] = le.fit_transform(y_train_imputed[col])
    y_test_imputed[col] = le.transform(y_test_imputed[col])
# Impute regression columns
reg_imputer = SimpleImputer(strategy='mean')
for idx in set(range(y_train.shape[1])) - set(classification_indices):
    col = y_train.columns[idx]
    y_train_imputed[col] = reg_imputer.fit_transform(y_train[[col]]).ravel()
    y_test_imputed[col] = reg_imputer.transform(y_test[[col]]).ravel()
```

```python
for i, target in enumerate(y_train_imputed.columns):
    y_col=y_train_imputed[target]
    
    if i in classification_indices:
        model=GradientBoostingClassifier(n_estimators=200,max_depth=4,learning_rate=0.07,subsample=0.9,random_state=42)
        scoring='accuracy'
    else:
        model = GradientBoostingRegressor(
        n_estimators=400,
        max_depth=5,
        learning_rate=0.05,subsample=0.85,
        min_samples_split=4,
        min_samples_leaf=2,    
        random_state=42)
        scoring='r2'
    
    pipeline = Pipeline([
        ('preprocessor',preprocessor),
        ('model',model)
    ])
    
    
    
    if i in classification_indices:
        cv=StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
        cv_scores=cross_val_score(pipeline,x_train,y_col,cv=cv,scoring=scoring)
        print(f"Cross-validated Accuracy for {target}: {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")

    else:
        cv = KFold(n_splits=5,shuffle=True,random_state=42)
        cv_scores = cross_val_score(pipeline, x_train, y_col, cv=cv, scoring=scoring)
        print(f"Cross-validated R2 for {target}: {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")
        
              
    pipeline.fit(x_train,y_col)
    y_pred=pipeline.predict(x_test)
                    
    if i in classification_indices:
        acc = accuracy_score(y_test_imputed[target], y_pred)
        print(f'Output column {72+i} ({target}) Accuracy: {acc:.4f}')  # Fixed
    else:
        mae = mean_absolute_error(y_test_imputed[target], y_pred)
        r2 = r2_score(y_test_imputed[target], y_pred)
        print(f'Output column {72+i} ({target}) MAE: {mae:.4f}, R2: {r2:.4f}')
              
    joblib.dump(pipeline, f'output_col_{72+i}_{target}_pipeline.pkl')
    models[f'output_col_{72+i}_{target}'] = pipeline          
    
```
